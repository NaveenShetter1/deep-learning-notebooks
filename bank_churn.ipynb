```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset=pd.read_csv("/content/sample_data/Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
# here 1st rownumber, customerid, surname will be not helpfull for finding output, so we will remove those columns
```

```python
x=dataset.iloc[:,3:-1]
```

```python
x.head()
```

```python
y=dataset.iloc[:,-1]
```

```python
y.head()
```

```python
# now we have categorical data like geography and gender
```

```python
# now by using label encoder we are going to convert categorical data to numerical data
```

```python
# gender column to numerical
```

```python
x.iloc[:,2].head()
```

```python
x.iloc[:,2].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

x.iloc[:,2]=le.fit_transform(x.iloc[:,2])
```

```python
x.iloc[:,2].unique()
```

```python
x.head()
```

```python
x
```

```python
# now to convert geography column from categorical to numerical
```

```python
x.iloc[:,1]
```

```python
x.iloc[:,1].unique()
```

```python
# using one hot encoder
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
```

```python
# all columns are filled with data
```

```python
# now split the into train and test
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# now here in deep learning feature scaling is very important
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
y_test=sc.transform(x_test)
```

```python
x_train[0]
```

```python
# now building our ANN
```

```python
# adding 1st hidden layer
```

```python
ann=tf.keras.models.Sequential()
```

```python
# now here sequential is the architecture of layers under which we will be having types like dense, RNN ext, now adding 1st hidden layer using dense layer
```

```python
ann.add(tf.keras.layers.Dense(6, activation='relu'))
```

```python
# now adding 2nd hidden input layer
```

```python
ann.add(tf.keras.layers.Dense(6, activation='relu'))
```

```python
# now adding output layer
```

```python
ann.add(tf.keras.layers.Dense(1, activation='sigmoid'))

# here activation is sigmoid because the output will be in binary either in 0 or 1
# if you the output is in multiclass than we will use activation as softmax
```

```python
# dense means fully connected layer, here ANN means architecture
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# here optimizer adam means it is using suckostic gd and here loss is binary_crossentropy because the activtion is binary means sigmoid
# we can give metrics multiple values under it
```

```python
# training the ANN by training data
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```
