```python
import os
import numpy as np
import pandas as pd
dataset=pd.read_csv('/content/Social_Network_Ads.csv')
dataset.head()
```

```python
# now convert gender data to category to numerical
```

```python
dataset['Gender']=dataset['Gender'].astype('category')
dataset['Gender']=dataset['Gender'].cat.codes
dataset.head()
```

```python
# we need to remove 1st column i.e user id so we are making
dataset=dataset.iloc[:,1:]
dataset.head()
```

```python
# to check which having missing data

dataset.isnull().sum()
```

```python
dataset
```

```python
# to seperate independent(x) and target (y)

x=dataset.iloc[:,:-1]
# all rows and all columns except last column

y=dataset.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
# we need to import train_test_split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

#Multilayer Perceptron Model or DNN or ANN

```python
import tensorflow as tf # tensorflow a backend language
from tensorflow import keras # keras a front-end language
from keras.layers import Dense #fully connected layer
from keras import Sequential
```

```python
model=Sequential()
# first hidden layer
model.add(Dense(128,activation='relu',input_dim=3))
# second hidden layer
model.add(Dense(128,activation='relu'))
# output
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=100)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc_train_x=sc.fit_transform(x_train)
sc_test_x=sc.transform(x_test)
```

```python
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
history=model1.fit(sc_train_x,y_train,validation_data=(sc_test_x,y_test),epochs=100)
```
